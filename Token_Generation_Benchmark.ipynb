```python
# Token_Generation_Benchmark.ipynb

import time
import matplotlib.pyplot as plt

def generate_tokens(model_name, tokens):
    start = time.time()
    if model_name == "Grok-3 (FP8)":
        time.sleep(tokens * 0.0005)
    else:
        time.sleep(tokens * 0.0012)
    end = time.time()
    return round(end - start, 3)

tokens_to_generate = [100, 500, 1000, 2000]
grok_times = [generate_tokens("Grok-3 (FP8)", t) for t in tokens_to_generate]
gpt4_times = [generate_tokens("GPT-4 (FP16)", t) for t in tokens_to_generate]

# Plot
plt.figure(figsize=(6, 4))
plt.plot(tokens_to_generate, grok_times, label="Grok-3 (FP8)", marker='o')
plt.plot(tokens_to_generate, gpt4_times, label="GPT-4 (FP16)", marker='x')
plt.title("Token Generation Time Benchmark")
plt.xlabel("Tokens")
plt.ylabel("Time (s)")
plt.legend()
plt.grid(True)
plt.show()
```
