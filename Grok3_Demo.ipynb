```python
# Grok3_Demo.ipynb

## üìå Grok-3 Inference Simulation Notebook

from time import time
import matplotlib.pyplot as plt
import numpy as np
import time as t

print("üîç Running Grok‚Äë3 Inference Simulation...")

# Simulate Inference Function
def simulate_inference(input_text, mode='FP8'):
    start = time()
    tokens_generated = len(input_text.split()) * 10
    if mode == 'FP8':
        t.sleep(0.5)  # Simulate fast FP8
    else:
        t.sleep(1.2)  # Simulate slower FP16/FP32
    end = time()
    return {
        'tokens_generated': tokens_generated,
        'time_taken': round(end - start, 2),
        'mode': mode
    }

# Run both simulations
result_fp8 = simulate_inference("Grok3 is an FP8 optimized model.", mode='FP8')
result_fp16 = simulate_inference("Grok3 is an FP8 optimized model.", mode='FP16')

print(result_fp8)
print(result_fp16)

# Visual comparison
modes = ['FP8', 'FP16']
times = [result_fp8['time_taken'], result_fp16['time_taken']]

plt.figure(figsize=(6, 4))
plt.bar(modes, times, color=['green', 'orange'])
plt.title('Inference Time Comparison')
plt.ylabel('Time (seconds)')
plt.show()
```
