```python
# MoE_Routing_Simulation.ipynb

import random
import matplotlib.pyplot as plt

print("ðŸ”€ Simulating MoE Routing for Grok-3")

experts = ["Expert A", "Expert B", "Expert C", "Expert D"]
token_inputs = [f"Token_{i}" for i in range(20)]

# Simulate routing decisions
routing_log = {expert: 0 for expert in experts}

for token in token_inputs:
    selected_expert = random.choices(experts, weights=[0.4, 0.3, 0.2, 0.1])[0]
    routing_log[selected_expert] += 1
    print(f"{token} routed to {selected_expert}")

# Plot
plt.figure(figsize=(6,4))
plt.bar(routing_log.keys(), routing_log.values(), color='purple')
plt.title("Token Distribution Across Experts (Grok-3 MoE Simulation)")
plt.ylabel("Number of Tokens")
plt.xlabel("Expert")
plt.grid(True)
plt.show()
```
